```python
info_list = []

# Loop through each movie item
for movie in movie_info:
    
    dict1 = {
        "name": "",
        "image": "",
        "certificate": "",
        "duration": "",
        "genre": "",
        "rating": "",
        "release_date": "",
        "summary": "",
        "votes": "",
        "gross": ""
    }
    
    try:
        name = movie.find("h3", class_="lister-item-header").find("a").text
        print(name)
        dict1.update(name = name)
        image = movie.find("img")["loadlate"]
        print(image)
        dict1.update(image = image)
        release_date = (movie.find("div", class_="list-description").find("p").text).split("\nStudio")[0]
        print(release_date)
        dict1.update(release_date = release_date)
        certificate = movie.find("span", class_="certificate").text
        print(certificate)
        dict1.update(certificate = certificate)
        genre = (movie.find("span", class_="genre").text).split("\n")[1]
        print(genre)
        dict1.update(genre = genre)
        duration = (movie.find("span", class_="runtime").text).split(" ")[0]
        print(duration)
        dict1.update(duration = duration)
        rating = movie.find("span", class_="ipl-rating-star__rating").text
        print(rating)
        dict1.update(rating = rating)

        
        # Find elements that contain info for votes and gross revenue
        items = movie.find_all("p", class_="text-muted text-small")
        # Loop through each of these items
        for item in items:
            # Find each specific element of information
            value = item.find_all("span", {"name": "nv"})
            # Loop through each one
            for i in range(len(value)):
                # Assign the text value to a variable
                value_text = value[i]["data-value"]
                print(value_text)
                # If the text value is not empty...
                #if (value_text != None):
                # If the text value's (nth + 1) position is even, assign it to the gross key
                if ((i +1) % 2 != 0):
                    votes = value_text
                    print(votes)
                    dict1.update(votes = votes)
                else:
                    gross = value_text.replace(",", "")
                    print(gross)
                    dict1.update(gross=gross)
                        

        # Append the dictionary to the list
        info_list.append(dict1)
        
    except (AttributeError):
        info_list.append(dict1)
        continue
#         print("no info")
    except(TypeError):
        info_list.append(dict1)
        continue
#         print("no info")
    
```
